```python
df.rename(columns={"CLASIFFICATION_FINAL":"CLASSIFICATION_FINAL"}, inplace=True)
```

```python
df.replace([97,98,99],np.nan,inplace=True)
```

```python
df["SEX"].replace({1:"female", 2:"male"},inplace=True)
df["PREGNANT"].replace({1:"yes", 2:"no"},inplace=True)
df["DIABETES"].replace({1:"yes", 2:"no"},inplace=True)
df["COPD"].replace({1:"yes", 2:"no"},inplace=True)
df["ASTHMA"].replace({1:"yes", 2:"no"},inplace=True)
df["INMSUPR"].replace({1:"yes", 2:"no"},inplace=True)
df["HIPERTENSION"].replace({1:"yes", 2:"no"},inplace=True)
df["CARDIOVASCULAR"].replace({1:"yes", 2:"no"},inplace=True)
df["RENAL_CHRONIC"].replace({1:"yes", 2:"no"},inplace=True)
df["OTHER_DISEASE"].replace({1:"yes", 2:"no"},inplace=True)
df["OBESITY"].replace({1:"yes", 2:"no"},inplace=True)
df["PATIENT_TYPE"].replace({1:"Return Home", 2:"Hospitalized"},inplace=True)
df["TOBACCO"].replace({1:"yes", 2:"no"},inplace=True)
df["ICU"].replace({1:"Yes", 2:"no"},inplace=True)
```
